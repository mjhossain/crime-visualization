**Importing Folium, Pandas, and SodaPy**

```python
import folium
import csv

import pandas as pd
from sodapy import Socrata

from folium import plugins
from folium.plugins import HeatMap
```

**Importing AirBnb Data and Creating Pandas Dataframe from it.**

```python
# Importing AirBnb Data (2019) and creating pandas data frame
airbnb_data = pd.read_csv('/content/gdrive/MyDrive/data/AB_NYC_2019.csv', delimiter=',', nrows=1500)
abnb_df = pd.DataFrame.from_records(airbnb_data)

# Preparing AirBnb Data for map usage
lat = abnb_df['latitude'].apply(lambda x: float(x))
lon = abnb_df['longitude'].apply(lambda x: float(x))
price = abnb_df['price'].apply(lambda x: str(x))
```

**Importing NYPD Crime Data and Creating a Pandas Dataframe.**

```python
client = Socrata("data.cityofnewyork.us", None)
results = client.get("qgea-i56i", limit=1000)
nypd_df = pd.DataFrame.from_records(results)
```

**Creating a Folium Map**

```python
# Creating Folium Map
map = folium.Map(location=[40.730610, -73.935242], tiles='Stamen Toner',zoom_start = 12)
```

Creating heatmap data from the NYPD Dataframe

```python
heat_data = [[row['latitude'],row['longitude']] for index, row in nypd_df.iterrows()]
HeatMap(heat_data).add_to(map)
```

**Creating Map locations AirBnb Data**

```python
for index in abnb_df.index:
  folium.Circle(
    radius=10,
    popup='$' + price[index] + ' per night',
    location=[lat[index], lon[index]],
    color="blue",
    fill=False,
    ).add_to(map)
```
